```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

## prepare data

```python
confirmed_df = pd.read_csv("covid19_confirmed_global.txt")

cols = confirmed_df.keys()
select_day_cols = [c for c in cols if c[:2] == "10" and c[-2:] == "21"]

confirmed_df = confirmed_df[["Country/Region", "Lat", "Long", *select_day_cols]]
confirmed_df["confirm_cases_oct_21"] = confirmed_df[select_day_cols].sum(axis=1)

data = confirmed_df[["Country/Region", "Lat", "Long", "confirm_cases_oct_21"]]

data = data[~data["Lat"].isnull()]
data.to_csv("world_oct_21_confirm_cases.csv", index=False)
```
